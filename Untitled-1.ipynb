```python
import pandas as pd
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.drop(['Cabin','Name','Ticket','Fare'],axis=1)
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['Sex'] = encoder.fit_transform(data['Sex'])
```

```python
data.head()
```

```python
one_hot_encoder = pd.get_dummies(data['Embarked'], prefix='Embarked')
data = pd.concat([data, one_hot_encoder], axis=1)
```

```python
data.head()
```

```python
data = data.drop(['Embarked'],axis=1)
```

```python
data.head()
```

```python
data['Embarked_C'] = encoder.fit_transform(data['Embarked_C'])
data['Embarked_Q'] = encoder.fit_transform(data['Embarked_Q'])
data['Embarked_S'] = encoder.fit_transform(data['Embarked_S'])
```

```python
data.head()
```

```python
data.dropna(axis=1, inplace=True)
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



# Create boxplots for each feature
plt.figure(figsize=(12, 6))  # Adjust the figure size
for i, column in enumerate(data.columns, 1):
    plt.subplot(1, len(data.columns), i)  # Create subplots
    sns.boxplot(y=data[column])
    plt.title(column)

plt.tight_layout()
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)

# Add title and labels
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
features= data.drop(['Survived'],axis=1)
```

```python
target = data.Survived
```

```python
test = pd.read_csv('test.csv')
test.head()
```

```python
test = test.drop(['Cabin','Name','Ticket'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
test['Sex'] = encoder.fit_transform(test['Sex'])
```

```python
one_hot_encoder = pd.get_dummies(test['Embarked'], prefix='Embarked')
test = pd.concat([test, one_hot_encoder], axis=1)
```

```python
test = test.drop(['Embarked'],axis=1)
```

```python
test['Embarked_C'] = encoder.fit_transform(test['Embarked_C'])
test['Embarked_Q'] = encoder.fit_transform(test['Embarked_Q'])
test['Embarked_S'] = encoder.fit_transform(test['Embarked_S'])
```

```python
test.dropna(axis=1, inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(features,target)
```

```python
model_prediction = model.predict(features)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Accuracy score
print("Accuracy:", accuracy_score(target, model_prediction))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(target, model_prediction))

# Classification Report
print("Classification Report:")
print(classification_report(target, model_prediction))
```

```python
from sklearn.svm import SVC
model_2 = SVC()
model_2.fit(features,target)
```

```python
model_prediction_2 = model_2.predict(features)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Accuracy score
print("Accuracy:", accuracy_score(target, model_prediction_2))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(target, model_prediction_2))

# Classification Report
print("Classification Report:")
print(classification_report(target, model_prediction_2))
```
